```python
import xml.etree.ElementTree as et
tree=et.parse('/home/robert/Pulpit/FS-17-01-2020-S.xml')
```

```python
elem=tree.getroot()
for p in elem.findall('Invoice-Lines/Line/Line-Item'):
    ean=p.find('EAN').text
    ilosc=p.find('InvoiceQuantity').text
    lp=p.find('Number').text
    zakup_netto=p.find('InvoiceUnitNetPrice').text
    detal=p.find('InvoiceUnitRetailPrice').text
    vat=p.find('TaxRate').text
    tytul=p.find('ItemDescription').text
    jm=p.find('UnitOfMeasure').text
    print(lp, tytul, ean, ilosc.replace('.', ','), zakup_netto.replace('.', ','), detal.replace('.', ','), vat, jm)
    
```
